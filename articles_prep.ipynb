```python
import pandas as pd
df_art1 = pd.read_csv('articles1.csv', sep=',')
df_art2 = pd.read_csv('articles2.csv', sep=',')
df_art3 = pd.read_csv('articles3.csv', sep=',')
```

```python
df_art1 = df_art1.drop(['author', 'date', 'year', 'month', 'url'], axis=1)
df_art2 = df_art2.drop(['author', 'date', 'year', 'month', 'url'], axis=1)
df_art3 = df_art3.drop(['author', 'date', 'year', 'month', 'url'], axis=1)
```

```python
sets = [df_art1, df_art2, df_art3]
df_art123 = pd.concat(sets)
```

```python
df_art123['publication'].value_counts()
```

```python
max = 4000

df = df_art123.groupby('publication').head(max)
```

```python
df = df[df['publication'] != 'Talking Points Memo']
df['publication'].value_counts()
```

```python
df['political_bias'] = 'Other' 
 
df.loc[df['publication'] == 'New York Times', 'political_bias'] = (-2.2)
df.loc[df['publication'] == 'Breitbart', 'political_bias'] = 5
df.loc[df['publication'] == 'CNN', 'political_bias'] = (-1.3)
df.loc[df['publication'] == 'Business Insider', 'political_bias'] = (-2.6)
df.loc[df['publication'] == 'Atlantic', 'political_bias'] = (-4)
df.loc[df['publication'] == 'Fox News', 'political_bias'] = 3.2
df.loc[df['publication'] == 'Buzzfeed News', 'political_bias'] = (-4)
df.loc[df['publication'] == 'National Review', 'political_bias'] = 2.5
df.loc[df['publication'] == 'New York Post', 'political_bias'] = 1.8
df.loc[df['publication'] == 'Guardian', 'political_bias'] = (-2.4)
df.loc[df['publication'] == 'NPR', 'political_bias'] = (-2)
df.loc[df['publication'] == 'Reuters', 'political_bias'] = (-0.7)
df.loc[df['publication'] == 'Vox', 'political_bias'] = (-3.5)
df.loc[df['publication'] == 'Washington Post', 'political_bias'] = (-2.2)
```

```python
df.to_csv('articles.csv', index=False, sep=',')
```
